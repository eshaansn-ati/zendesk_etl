```python
import json

file_path = "../data/transform/fields/ticket_fields_1764932460.json"
with open(file_path, "r", encoding="utf-8") as f:
    fields = json.load(f)
field_infos = [
    (field.get("title"), field.get("active"))
    for field in fields
    if "title" in field and "active" in field
]
print("Field names (with active status):")
for name, active in field_infos:
    print(f"{name} (active: {active})")
```

```python
org_map_file = '/home/eshaan/Projects/analytics/zendesk_pipeline/data/transform/organizations/organization_map_1765260441.json'
organization_map = {}
if org_map_file:
    with open(org_map_file, 'r', encoding='utf-8') as f:
        organization_map = json.load(f)
    print(f"Loaded organization map from {org_map_file} ({len(organization_map)} organizations)")
else:
    print("⚠️  Warning: No organization map found. Organization IDs will not be joined with names.")
```
